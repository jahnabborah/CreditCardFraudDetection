Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas df
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
#last 5 rows of the dataset
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0.0-->normal transaction
1.0-->fraudulent transaction

```python
#seperating the data for analysis
legit=credit_card_data[credit_card_data.Class==0.0]
fraud=credit_card_data[credit_card_data.Class==1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling...
Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions.
Number of fraudulent transactions-204

```python
legit_sample=legit.sample(n=204)
```

Concatenating two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets...




(Features are nothing ,but rows..and Targets are columns)

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic regression(it is used for binary classification problems)

```python
model=LogisticRegression(max_iter=1000)
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model evaluation.
(Accuracy score)

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on Test data:',test_data_accuracy)
```
